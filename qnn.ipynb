```python
# Import Libraries
from qiskit import QuantumCircuit, transpile, Aer, assemble
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
x0 = 2
x1 = 4
x2 = 7
```

```python
#Inisialisasi kuantum sirkuit
qc = QuantumCircuit(3)

#Lapisan variasional encode
for i,x in enumerate([x0,x1,x2]):
    qc.h(i)
    qc.ry(x,i)

qc.barrier()

#Lapisan parameterize circuit (PC)
for qubit in range(3):
    qc.rx(0.3*x, qubit)
    qc.rz(0.2*x, qubit)
    if qubit<2:
        qc.cx(qubit, qubit+1)
#Measurement
qc.measure_all()

#Tampil sirkuit
qc.draw(output="mpl")
```
