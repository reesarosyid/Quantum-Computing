## Qubit Ganda Q diperoleh dari dua Qubit tunggal q1 dan q2

```python
q1 = Statevector([1/sqrt(5), 2/sqrt(5)]) # Prob |0> = 1/5      prob |1> = 4/5
q2 = Statevector([1/sqrt(2), 1/sqrt(2)]) # Prob |0> = 1/2      prob |1> = 1/2

print('Qubit tunggal q1 dan q2')
display(q1.draw('latex'))
display(q2.draw('latex'))

print()
print('Qubit Ganda Q = q1âŠ—q2 ')
Q = q1.tensor(q2)  # `(q1âŠ—q2)
display(Q.draw('latex'))
display(plot_bloch_multivector(Q))
```

```python
probs = Q.probabilities()
statistics = Q.sample_counts(1000)
display(statistics)
print('Pengukuran 1000 kali')
print('probabilitas status |00> = ',np.round(probs[0],1), '   Jumlah kemunculan :',statistics['00'])
print('probabilitas status |01> = ',np.round(probs[1],1), '   Jumlah kemunculan :',statistics['01'])
print('probabilitas status |10> = ',np.round(probs[2],1), '   Jumlah kemunculan :',statistics['10'])
print('probabilitas status |11> = ',np.round(probs[3],1), '   Jumlah kemunculan :',statistics['11'])
plot_histogram(statistics)
```
