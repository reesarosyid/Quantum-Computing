```python
import matplotlib.pyplot as plt
import numpy as np

from numpy import pi
from PIL import Image

from qiskit import QuantumCircuit, Aer, execute
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes 
from qiskit.visualization import circuit_drawer
```

```python
img = Image.open('fig1.png')
plt.figure(figsize=(20, 10))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
num_qubits = 4
qc         = QuantumCircuit(num_qubits)

# Encoding
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)
qc.rx(np.pi/2, 0)
qc.rx(np.pi/2, 1)
qc.rx(np.pi/2, 2)
qc.rx(np.pi/2, 3)

qc.barrier()

# Ansatz
qc.ry(np.pi/2, 0)
qc.rz(np.pi/2, 0)
qc.ry(np.pi/2, 1)
qc.rz(np.pi/2, 1)
qc.ry(np.pi/2, 2)
qc.rz(np.pi/2, 2)
qc.ry(np.pi/2, 3)
qc.rz(np.pi/2, 3)
qc.cx(0, 1)
qc.cx(1, 2)
qc.cx(2, 3)
qc.cx(3, 0)

# Decoding
qc.measure_all()

qc.draw(output='mpl', style="clifford")
```

```python
num_qubits = 4
qc         = QuantumCircuit(num_qubits)

# Encoding
qc.h([0,1,2,3])
qc.rx(np.pi/2, [0,1,2,3])

qc.barrier()

# Ansatz
qc.ry(np.pi/2, [0,1,2,3])
qc.rz(np.pi/2, [0,1,2,3])
qc.cx([0,1,2,3], [1,2,3,0])

# Decoding
qc.measure_all()

qc.draw(output='mpl', style="clifford")
```

```python
num_qubits = 4
qc         = QuantumCircuit(num_qubits)

# Encoding
for i in range (num_qubits):
    qc.h(i)
    qc.rx(np.pi/2, i)

qc.barrier()

# Ansatz
for i in range (num_qubits):
    qc.ry(np.pi/2, i)
    qc.rz(np.pi/2, i)
for i in range(num_qubits-1): 
    qc.cx(i, i+1)
qc.cx(3, 0)

# Decoding
qc.measure_all()

qc.draw(output='mpl', style="clifford")
```
