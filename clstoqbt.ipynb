```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_bloch_multivector

# Fungsi untuk mengonversi nilai fitur biner menjadi keadaan kuantum
def classical_to_quantum(feature_values):
    n = len(feature_values)
    qc = QuantumCircuit(n, n)
    
    for i, value in enumerate(feature_values):
        if value == 1:
            qc.x(i)
    
    return qc

# Contoh fitur klasik
classical_features = [1, 0, 1]

# Konversi fitur klasik ke keadaan kuantum
quantum_circuit = classical_to_quantum(classical_features)

# Visualisasi sirkuit kuantum
print(quantum_circuit)

# Simulasi sirkuit kuantum
simulator = Aer.get_backend('statevector_simulator')
job = assemble(transpile(quantum_circuit, simulator))
result = simulator.run(job).result()

# Visualisasi hasil simulasi menggunakan Bloch vector plot
statevector = result.get_statevector()
plot_bloch_multivector(statevector)
```

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble, execute
from qiskit.visualization import plot_bloch_multivector

x = 10

qc1 = QuantumCircuit(1)

qc1.h(0)

simulator = Aer.get_backend("statevector_simulator")
result = execute(qc1, simulator).result()
statevector = result.get_statevector()

print(qc1)

print(statevector)

print(qc1.decompose().draw(output="mpl", fold=50))
plot_bloch_multivector(statevector)   
```

```python
# Membuat QC dengan 1 qubit 
qc2 = QuantumCircuit(1)

# Rotasi X pada qubit dengan nilai fitur x sebagai sudut rotasi
qc2.rx(x, 0)

simulator = Aer.get_backend('statevector_simulator')
result = execute(qc2, simulator).result()
statevector = result.get_statevector()
print(qc2)
print(statevector)

print(qc1.decompose().draw(output="mpl", fold=50))
plot_bloch_multivector(statevector)  
```
