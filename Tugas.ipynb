```python
# Import Libraries
from qiskit import QuantumCircuit, transpile, QuantumRegister, Aer, execute, ClassicalRegister, assemble
from qiskit.tools.visualization import plot_histogram
from qiskit.quantum_info import Statevector
```

```python
qr = QuantumRegister(3, 'q')
cr = ClassicalRegister(3, 'c')

qc = QuantumCircuit(qr,cr)

qc.h(qr[0])
qc.h(qr[2])
qc.x(qr[1])
qc.barrier()
qc.cx(qr[0],qr[1])
qc.z(qr[2])
qc.z(qr[0])
qc.barrier()
qc.z(qr[0])
qc.z(qr[2])
qc.cx(qr[0],qr[1])
qc.barrier()
qc.h(qr[0])
qc.x(qr[1])
qc.h(qr[2])
qc.measure(qr[0],cr[0])
qc.measure(qr[1],cr[1])
qc.measure(qr[2],cr[2])
# Menggambar sirkuit kuantum
qc.draw('mpl')
```

```python
simulator = Aer.get_backend("aer_simulator") # simulasi kuantum dilakukan di komputer lokal
qc.save_statevector()
tqc = transpile(qc, simulator)
job = simulator.run(tqc)
result = job.result()
statevector = result.get_statevector(tqc, 4)
print(statevector)
```

```python
# Use Aer's qasm_simulator
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc, simulator, shots=1000)
# Return counts
counts = job.result().get_counts(qc)
print(counts)
plot_histogram(counts)
```
